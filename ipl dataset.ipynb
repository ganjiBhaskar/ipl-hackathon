### Most number of wickets taken by a bowler is


```python
def dismissal_kind(x):
    if (x=="caught")|(x=="caught")|(x=="lbw")|(x=="caught and bowled")|(x=="stumped")|(x=="hit wicket"):
        return 1
    
    else:
        return 0
```

```python
df2["Wickets"]=df2["dismissal_kind"].apply(dismissal_kind)
```

```python
df3=df2.groupby(by="bowler")["Wickets"].sum().reset_index(name='wickets')

```

```python
df3.sort_values(by='wickets')
```

**from the above the A Mishra has taken the more no of wickets**
